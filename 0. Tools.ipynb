### Numpy Random Functions and Generators

```python
n=1000
k=2
x=np.random.random_sample((n,k)) # nxk matrix of random numbers <- [0,1] 
```

### Plots

```python
fig, [ax1, ax2] = plt.subplots(1, 2, sharex="col", sharey="row", figsize=(6,5))

# label in plot function shows up as legend after calling ax.legend() or plt.legend() function
ax1.plot(x, label='x')
h = ax2.hist(x, bins=25)

# Setting axis labels on plt object
plt.ylabel('Number of outcomes')
plt.xlabel('Value of X')
plt.title('Histogram of X = sum of %s random events'%(k,))

# Setting limits on plt object
plt.xlim([-0.25, 1.5])
plt.ylim([-0.25, 1.5])

# Setting axis labels for each subplot
ax1.set_ylabel('p')
ax1.set_xlabel('N - Class Size')
ax1.set_title('exactly_2_students_have_the_same_birthday')
ax1.legend()
```
