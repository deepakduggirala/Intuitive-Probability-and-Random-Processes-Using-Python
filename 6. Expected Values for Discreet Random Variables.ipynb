## Expected Value of a Discreet Random Variable

$E(X) = \sum_{i}^{} x_i pX[x_i]$ 

$x_i$ values of the random variable X

$pX$ - PMF od random variable X

### Interpretations
- It is the outcome of an experiment in the long run
- Average of the outcome of the experiment repeated many times
- Center of Mass of linearly arranged masses where $x_i$ is the mass and its position is given by $pX[x_i]$
- Expected value is the best predictor of the outcome of the experiment, where best means this value minimizes the Mean Square Error (MSE)


### Properties
- It does not indicate the most probable value of the random variable
- Different PMF can have smae expected value
- Not all PMFs have expected value, for some cases the the sum goes to infinity


## Expected Value for a function of Random Variable

Let $Y = g(X)$ is the mapping of random variable X

$E(Y) = \sum_{i}^{}y_i PY(y_i)$

$E(Y) = \sum_{i}^{}g(x_i) PY(g(x_i))$

$E(Y) = \sum_{i}^{}g(x_i) PX(x_i)$ since $PX = PY . g$

$E(g[X]) = \sum_{i}^{}g(x_i) PX(x_i)$


### Expectation Operator is linear

$E(a_1X_1 + a_2X2) = a_1E(X_1) + a_2E(X_2)$

## Variance and Moments of a Random Variable

$Var(X) = E[(X - E[X])^2]$

Variance measures the average squared deviation from the mean of a random variable

$Var(X) = E[g(X)]$ where $g(X) = (X-E[X])^2$

$Var(X) = \sum_{i}^{} (x_i-E[X])^2 PX(x_i)$ applying $E(g[X]) = \sum_{i}^{}g(x_i) PX(x_i)$

#### Alternate Expression for Variance

$Var(X) = E[(X - E[X])^2]$

$Var(X) = E[X^2 - 2XE[X] + E[X]^2$  expanding

$Var(X) = E[X^2] - 2E[X]E[X] + E[X]^2$  applying linearity of E and E[constant] = constant

$Var(X) = E[X^2] - E[X]^2$

#### Properties

$var(c) = 0$

$var(X + c) = var(X)$

$var(cX) = c^2var(X)$

Variance is a non-linear operator

Variance is the minimum value of MSE or the value of MSE at x = E(X), since E(X) minimizes MSE.



**Variance limits how well we can predict the outcome of an experiment.**

**Expected value is the best prediction but variance qualifies how good the prediction is**



### First and Second Moments

$E(X)$ - First Moment

$E(X^2)$ - Second Moment

$E(X^n)$ - nth moment

#### Properties

- If nth moment is finite then 1 to n-1 moments are also finite
- Mean and Variance of a random variable will exist if second moment is finite
